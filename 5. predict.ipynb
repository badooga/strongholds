```python
import strongholds
import strongholds.graphing as mplt

import numpy as np

from matplotlib import pyplot as plt
import ipywidgets

fixed_rng = np.random.default_rng(0)
```

```python
heatmap = strongholds.generation_heatmap(2*10**5, ring_nums=range(3), rng=fixed_rng, concatenate=False)

grid = strongholds.generation_grid(range(3))
```

```python
predict = strongholds.Predict(grid=grid, heatmap=heatmap, rng=fixed_rng)
predict.add_throw(strongholds.Coordinates.from_polar(2500, np.pi/4), 80, 0.1)
predict.cumulative_probs.view(threshold=1e-3)
```

```python
%matplotlib widget
fig, ax = mplt.xz_subplots()

plots = predict.plot_throws(fig, ax)
```
