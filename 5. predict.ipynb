```python
import strongholds
import strongholds.graphing as mplt

import numpy as np

from matplotlib import pyplot as plt
import ipywidgets

fixed_rng = np.random.default_rng(0)
```

```python
heatmap = strongholds.generation_heatmap(2*10**5, ring_nums=range(3), rng=fixed_rng, concatenate=False)

grid = strongholds.generation_grid(range(3), center=True)
```

Below is an example from the API [example output](https://github.com/Ninjabrain1/Ninjabrain-Bot/wiki/API#example-output).

```python
predict1 = strongholds.Predict(grid=grid, heatmap=heatmap, rng=fixed_rng)
predict1.add_throw(strongholds.MCoordinates.from_rect(1199.63, -139.09), -161.13926034190884, 0.01)
predict1.cumulative_probs.view(chunk=True)
```

Below is an example from [this video](https://www.youtube.com/watch?v=ANgU7wxQE9M&t=460). Notice how the top three results here are the same as those from before the angle correction step in the video.

```python
predict2 = strongholds.Predict(grid=grid, heatmap=heatmap, rng=fixed_rng)
predict2.add_throw(strongholds.MCoordinates.from_rect(2315.30, 271.70), 24.25, 2.5e-3)
predict2.cumulative_probs.view(chunk=True)
```

Below is an example from [this video](https://www.youtube.com/watch?v=Xfjz4PblRao&t=344).

```python
predict3 = strongholds.Predict(grid=grid, heatmap=heatmap, rng=fixed_rng)
predict3.add_throw(strongholds.MCoordinates.from_rect(-770.70, 1401.30), 89.40, 5e-3)
predict3.cumulative_probs.view(chunk=True)
```

```python
%matplotlib widget
fig2, ax2 = mplt.xz_subplots()

plots = predict3.plot_throws(fig2, ax2)
```

Below is an example from the [example response](https://github.com/Ninjabrain1/Ninjabrain-Bot/wiki/API#example-output) for the stronghold API.
