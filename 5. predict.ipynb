```python
import strongholds
import strongholds.graphing as mplt

import numpy as np

from matplotlib import pyplot as plt
import ipywidgets

fixed_rng = np.random.default_rng(0)
```

```python
heatmap = strongholds.generation_heatmap(2*10**5, ring_nums=range(3), rng=fixed_rng, concatenate=False)

grid = strongholds.generation_grid(range(3), center=False)
```

```python
predict1 = strongholds.Predict(grid=grid, heatmap=heatmap, rng=fixed_rng)
predict1.add_throw(strongholds.Coordinates.from_rect(1199.63, -139.09), -161.13926034190884, 0.015)
predict1.cumulative_probs.view(chunk=True)
```

```python
predict2 = strongholds.Predict(grid=grid, heatmap=heatmap, rng=fixed_rng)
predict2.add_throw(strongholds.Coordinates.from_rect(85.65, 696.62), -29.83, 0.01)
predict2.cumulative_probs.view(chunk=True)
```

```python
predict3 = strongholds.Predict(grid=grid, heatmap=heatmap, rng=fixed_rng)
predict3.add_throw(strongholds.Coordinates.from_rect(-770.7, 1401.3), 89.4, 0.02)
predict3.cumulative_probs.view(chunk=True)
```

```python
p = strongholds.Coordinates.from_rect(-770.7, 1401.3)
s = strongholds.Coordinates.from_chunk(-137, 88, center=True)
(p - s).yrot
```

```python
%matplotlib widget
fig2, ax2 = mplt.xz_subplots()

ax2.scatter(s.x, s.z)

plots = predict3.plot_throws(fig2, ax2)
```
